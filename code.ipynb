```python
import numpy as np
import pandas as pd
```

```python
#Import the data set
movies = pd.read_csv('Tmdb_5000_movies.csv')
credits= pd.read_csv('Tmdb_5000_credits.csv')
movies.head()
```

```python
credits.head()
```

```python
#merging the dataset
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#only those columns are used which has a significant difference in the dataset
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#converting it into ['action','adventure','ffantacy','sci-fi']
```

```python
#used for convering string to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')  
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#combining all the columns
movies['tags']=movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#converting text to lower case 
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python


new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
#first movie with each words count
vectors[0] 
```

```python
#nltk is a library in nlp used for stemming 
import nltk
```

```python
#stemming is used to reduce the set of words having same meaning
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
ps.stem('loving')
```

```python
#now since some words are reapting like actor and actors which has the same meaning so removing them using a technique called stemming
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
#In sklearn we have a library for cosine similarity between vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity =cosine_similarity(vectors)
```

```python
#first movies similarity with every other movie
similarity[0]
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index=new_df[new_df['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
